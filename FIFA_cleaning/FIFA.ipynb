```python
# Importing necessary Libraries
import pandas as pd
import numpy as np
from datetime import datetime

# Reading the data into the dataframe
data = pd.read_csv("dataset_used/fifa21.csv", low_memory=False)

# Viewing the entire dataframe columns at once
pd.set_option('display.max_columns', None)

# Viewing the top 5 items of the dataframe
data.head()
```

```python
# Checking the numbers of rows and columns in the dataframe
rows, columns = data.shape
print(f"Numbers of rows - {data.shape[0]}\nNumbers of columns - {data.shape[1]}")
```

```python
# Replacing " " (spaces) with "_" (an underscore)
data.columns = [column.replace(" ", "_").lower() for column in data.columns]

# Renaming the 'ova' column
data.rename(columns = {"↓ova": "ova"}, inplace=True)
data.columns
```

```python
# This particular block of code checks if there are null values in the dataframe
b = data.isnull().sum().to_list()
d = sum(b)
if d == 0:
    print("_DataFrame contains no NaN values_")
else:
    print(f"{d} NaN values exists")
```

```python
# Getting the generaal idea of how the dataframe looks like
data.info()
```

```python
# Going through the sample values of all the columns to be able to properly map out a plan to clean each of them
start = datetime.now()
for i in data.columns:
    print(f'{i.upper()} - \n{data[i].sample(n=4, random_state=4)}\n')
end = datetime.now()

print(f"This block of code took {end-start} seconds to run")
```

```python
# After careful consideration of the samples of each columns we can say that these are the columns where
# all of the dirty values lies
dirty = ["hits", "ir", "sm", "w/f", "club", "contract", "release_clause", "wage", "height",
         "weight", "joined", "loan_date_end", "value"]
data[dirty].head()
```

```python
# We start with the "hits" column
# Clean the hits column by removing the 'K' and do the multiplication.
data['hits'] = data['hits'].apply(lambda x: pd.to_numeric(x[:-1], errors='coerce') * 1000 if "K" in str(x) else x)

# There are null values so we fillna with 0.
data['hits'].fillna("0", inplace=True)
# The purpose of replacing NaN with 0 is that there is a high possiblity that some select players has no single hits
# If they were not played. There are players built for the bench lol.

# Convert the column to a data type of 'int64'
data.hits = data['hits'].astype('int64')
data['hits'].sample(n=10, random_state=4)
```

```python
# We move on to the columns that contains a '★'
# That is column 'ir', 'sm' and 'w/f'.
stars = dirty[1:4]

data[stars] = data[stars].apply(lambda x: x.str.replace(" ", ""))

for i in data[stars]:
    data[i] = data[i].apply(lambda x: int(x[0]) * x[1])

# This should look more presentable. :)
data[stars].sample(n=10, random_state=4)
```

```python
# Club next
data.club = data['club'].apply(lambda x: x.replace('\n', ''))

# Looks better.
data['club'].sample(n=10, random_state=4)
```

```python
# We will move on to the contract column.

# Adjusting the 'contract' column to take only useful values while ignoring '~' and other words.
data['contract_starts'] = data['contract'].apply(lambda x: x[-7:] if "On Loan" in x else x[:4].strip())

# Creating a new column called 'contract_ends' and taking values from the 'contract column' to populate it
data['contract_ends'] = data['contract'].apply(lambda x: x[:-7] if "On Loan" in x else x[:] if "Free" in x else x[-4:])

# Removing the 'contract' item from the dirty list
dirty.remove('contract')

# Adding the two new items we created to the dirty data list
dirty.extend(['contract_starts', 'contract_ends'])

# This should look better already :)
data[['contract_starts', 'contract_ends']].sample(n=10, random_state=4)
```

```python
# In the block we will be cleaning every column that is money related based on the data provided there is a distinct
# similarity between the columns that is they either have "K" for thousands or "M" for millions or are just hundreds
# and they all have the '€' symbol.
# Taking that into account we will create a somewhat lengthy lambda function to deal with it, that is remove the '€' sign
# and convert them to their respective denominations (if that's the correct word) :)
# Then after that we convert each of the columns to a floating number.

money_related = ["release_clause", "wage", "value"]
for i in data[money_related]:
    # Removed the '€' symbol
    data[i] = data[i].apply(lambda x: x.replace('€', ""))
    # Did the conversions here
    data[i] = data[i].apply(lambda x: pd.to_numeric(x[:-1], errors='coerce') * 1000 if "K" in x 
                            else pd.to_numeric(x[:-1], errors='coerce') * 1000000 if "M" in x 
                            else pd.to_numeric(x, errors='coerce'))
    data[i] = data[i].astype('int64')

# This should look better already :)
data[money_related].sample(n=10, random_state=4)
```

```python
# We go on to the 'weight' column

# We convert the 'ibs' to 'kg' while removing the 'ibs' and 'kg' unit of measure leaving it as a bare number.
data['weight(KG)'] = data['weight'].apply(lambda x: x[:-2] if "kg" in x
                                          else round(pd.to_numeric(x[:-3], errors='coerce') * 0.454, 2))

# Converting the column to floating point numbers
data['weight(KG)'] = data['weight(KG)'].astype('float64')

# Adding the new item we created to the dirty data list
dirty.append('weight(KG)')
# Dropping the weight column from the dataset and adding the weight(KG) to the dirty data list
dirty.remove('weight')

# Converted to KG looks good too :)
data['weight(KG)'].sample(n=10, random_state=4)
```

```python
# Replacing irrelevant values with the correct one
data['height'] = data['height'].apply(lambda x: x.replace('"', ""))
data['height'] = data['height'].apply(lambda x: x.replace("'", "."))

# lambda function to convert all to centimeter
data['height(cm)'] = data['height'].apply(lambda x: x[:-2] if "cm" in x
                                          else round(pd.to_numeric(x, errors='coerce') * 30.48, 2))

# Converting the 'height(cm)' column to floating point numbers
data['height(cm)'] = data['height(cm)'].astype('float64')

# Adding the new item we created to the dirty data list
dirty.append('height(cm)')
# Dropping the weight column from the dataset and adding the weight(KG) to the dirty data list
dirty.remove('height')

# Converted to cm looks good too :)
data['height(cm)'].sample(n=10, random_state=4)
```

```python
# Now we move to the loan colunm
data.loan_date_end = data['loan_date_end'].apply(lambda x: "Not On Loan" if x is np.nan else "On Loan").to_list()
data.rename(columns = {"loan_date_end": "loan_status"}, inplace=True)

# Adding the new item we created to the dirty data list
dirty.append('loan_status')
# Dropping the loan_date_end column from the dataset and adding the loan_status to the dirty data list
dirty.remove('loan_date_end')

# Looks better? Read text below for explanation 
data["loan_status"].sample(n=10, random_state=4)
```

The purpose of doing this particular conversion and introducing <b>Not On Loan</b> and <b>On Loan</b> values into this column is because when I checked the <b>loan_date_end</b> column and the <b>contract</b> column what I noticed was that the individuals <b>On Loan</b> and the dates attached to them is identical to the dates and the individuals in the <b>loan_date_end</b> so I felt it would be better to just create categorical values to identify those <b>On Loan</b> and those <b>Not On Loan</b>.

```python
# After working on the dirty data list this is what we have now.
# Clean data
data[dirty].head()
```

When I read the columns description above, we found out that certain columns are the combinations of several independent columns and they look like outliers but we can actually convert them to percentages because we already know what columns they sumed up to fill those particular colunms.

```python
# We have columns that are a combination of multiple columns for example:
# The attacking column is a combinaiton of crossing, finishing, heading, accuracy, short passing and volleys
# Knowing that each of the combination has a highest vallue of 100 we can deduce that to get the percentage of the 
# Attacking column we can apply the bsic percentage conversion formular
# attacking_value / total_of _highest_combo * 100
# That will be:

data['attacking(%)'] = data.attacking.apply(lambda x: round((x / 600) * 100), 2)
data['attacking(%)'].sample(n=10, random_state=4)
```

```python
# The skills column consists of the combination of the following; dribbling, curve, FK accuracy, long passing, ball control
# For the skills section we can also apply basic percentage conversion formular;
# skills_value / total_of _highest_combo * 100
# That will be:

data['skill(%)'] = data.skill.apply(lambda x: round((x / 500) * 100), 2)
data['skill(%)'].sample(n=10, random_state=4)
```

```python
# The movement column is also a combination of the following columns; acceleration,sprint speed, agility, reaction, balance
# For the movement column we will apply the basic percentage conversion formular;
# movement_value / total_of _highest_combo * 100
# That will be:

data['movement(%)'] = data.movement.apply(lambda x: round((x/500) * 100), 2)
data['movement(%)'].sample(n=10, random_state=4)
```

```python
# Power: Various power attributes of the player, including: Shot Power, Jumping, Stamina, Strength, Long Shots
# For the power column we will also apply the basic percentage conversion formular:
# power_value / total_of _highest_combo * 100
# That will be:

data['power(%)'] = data.power.apply(lambda x: round((x/500) * 100), 2)
data['power(%)'].sample(n=10, random_state=4)
```

```python
# Mentality: Various mentality attributes of the player, including:
# Aggression, Interceptions, Positioning, Vision, Penalties, Composure
# For the power column we will also apply the basic percentage conversion formular:
# mentality_value / total_of _highest_combo * 100
# That will be:

data['mentality(%)'] = data.mentality.apply(lambda x: round((x/500) * 100), 2)
data['mentality(%)'].sample(n=10, random_state=4)
```

```python
# Defending: Various defending attributes of the player, including Marking, Standing Tackle, Sliding Tackle
# For the defending column we will also apply the basic percentage conversion formular:
# defending_value / total_of _highest_combo * 100
# That will be:

data['defending(%)'] = data.defending.apply(lambda x: round((x/500) * 100), 2)
data['defending(%)'].sample(n=10, random_state=4)
```

```python
# Goalkeeping: Various goalkeeping attributes of the player, including:
# GK Diving, GK Handling, GK Kicking, GK Positioning, GK Reflexes
# For the goalkeeping column we will also apply the basic percentage conversion formular:
# goalkeeping_value / total_of _highest_combo * 100
# That will be:

data['goalkeeping(%)'] = data.goalkeeping.apply(lambda x: round((x/500) * 100), 2)
data['goalkeeping(%)'].sample(n=10, random_state=4)
```

<b>We are done cleaning</b><br>
So we will drop the colunms we don't need and that should be it.

```python
# We dropping all irrelevant columns because work has been done on them and new columns have been created to replace them.
col_to_drop = ['weight', 'height', 'attacking', 'defending', 'goalkeeping', 'power', 'mentality', 'contract', 'movement',
               'skill']

# for loop to drop each of them.
for i in col_to_drop:
    data.drop([i], axis=1, inplace=True)
```
